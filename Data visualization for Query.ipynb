```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3

# Set style for better-looking plots
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# ========================================
# STEP 1: CREATE IN-MEMORY DATABASE
# ========================================
print("Creating database...")
conn = sqlite3.connect(':memory:')

# ========================================
# STEP 2: LOAD ALL CSV FILES INTO DATABASE
# ========================================
print("Loading CSV files into database...")

# Load actor table
actor_df = pd.read_csv('actor.csv')
actor_df.to_sql('actor', conn, index=False, if_exists='replace')
print(f"âœ“ Loaded 'actor' table with {len(actor_df)} rows")

# Load content table
content_df = pd.read_csv('content.csv')
content_df.to_sql('content', conn, index=False, if_exists='replace')
print(f"âœ“ Loaded 'content' table with {len(content_df)} rows")

# Load ratings table
ratings_df = pd.read_csv('ratings.csv')
ratings_df.to_sql('ratings', conn, index=False, if_exists='replace')
print(f"âœ“ Loaded 'ratings' table with {len(ratings_df)} rows")

# Load actor_has_content table
actor_has_content_df = pd.read_csv('actor_has_content.csv')
actor_has_content_df.to_sql('actor_has_content', conn, index=False, if_exists='replace')
print(f"âœ“ Loaded 'actor_has_content' table with {len(actor_has_content_df)} rows")

# Load subscriber table
subscriber_df = pd.read_csv('subscriber.csv')
subscriber_df.to_sql('subscriber', conn, index=False, if_exists='replace')
print(f"âœ“ Loaded 'subscriber' table with {len(subscriber_df)} rows")

# Load subscriber_has_content table
subscriber_has_content_df = pd.read_csv('subscriber_has_content.csv')
subscriber_has_content_df.to_sql('subscriber_has_content', conn, index=False, if_exists='replace')
print(f"âœ“ Loaded 'subscriber_has_content' table with {len(subscriber_has_content_df)} rows")

print("\nâœ“ All tables loaded successfully!\n")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import mysql.connector

# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)

print("="*60)
print("EXECUTING SQL QUERIES AND CREATING VISUALIZATIONS")
print("="*60)

# ========================================
# CONNECT TO MySQL DATABASE
# ========================================
def get_connection():
    """Create MySQL connection"""
    return mysql.connector.connect(
        host='127.0.0.1',
        user='root',
        password='Cullen2121@',
        database='new_schema1'
    )

# ========================================
# SMART VISUALIZATION FUNCTION
# ========================================
def create_smart_visualization(df, query_num, title=None):
    """
    Automatically choose the best visualization based on data characteristics
    """
    if df is None or len(df) == 0:
        print("No data to visualize")
        return
    
    # Analyze the dataframe
    num_rows = len(df)
    num_cols = len(df.columns)
    
    # Get numeric columns
    numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
    text_cols = df.select_dtypes(include=['object']).columns.tolist()
    
    print(f"\nðŸ“Š Creating visualization for Query-{query_num}...")
    
    # Determine best visualization
    if num_cols == 2 and len(numeric_cols) == 1 and len(text_cols) == 1:
        # Case: One text column + one numeric column (most common)
        x_col = text_cols[0]
        y_col = numeric_cols[0]
        
        # Create label column if multiple text columns can be combined
        if 'Fname' in df.columns and 'Lname' in df.columns:
            df['Label'] = df['Fname'] + ' ' + df['Lname']
            x_col = 'Label'
        
        # Choose between horizontal/vertical bar based on number of items
        if num_rows <= 10:
            # Vertical bar for few items
            plt.figure(figsize=(12, 6))
            plt.bar(range(num_rows), df[y_col], color='steelblue', edgecolor='black', linewidth=1.2)
            plt.xticks(range(num_rows), df[x_col], rotation=45, ha='right')
            plt.ylabel(y_col.replace('_', ' ').title(), fontsize=12)
            plt.xlabel(x_col.replace('_', ' ').title(), fontsize=12)
        else:
            # Horizontal bar for many items (easier to read)
            plt.figure(figsize=(12, 10))
            plt.barh(range(num_rows), df[y_col], color='coral', edgecolor='black', linewidth=1.2)
            plt.yticks(range(num_rows), df[x_col])
            plt.xlabel(y_col.replace('_', ' ').title(), fontsize=12)
            plt.ylabel(x_col.replace('_', ' ').title(), fontsize=12)
            plt.gca().invert_yaxis()
        
        # Add value labels
        if num_rows <= 10:
            for i, v in enumerate(df[y_col]):
                plt.text(i, v + max(df[y_col])*0.01, str(v), ha='center', va='bottom', fontweight='bold')
        else:
            for i, v in enumerate(df[y_col]):
                plt.text(v + max(df[y_col])*0.01, i, str(v), va='center', fontweight='bold')
        
        plt.grid(alpha=0.3, linestyle='--')
    
    elif num_cols >= 3 and len(numeric_cols) >= 2:
        # Case: Multiple numeric columns - grouped bar chart
        plt.figure(figsize=(14, 7))
        
        x = range(num_rows)
        width = 0.8 / len(numeric_cols)
        
        for idx, col in enumerate(numeric_cols):
            offset = width * idx - (width * len(numeric_cols) / 2) + width / 2
            plt.bar([i + offset for i in x], df[col], width, label=col.replace('_', ' ').title())
        
        if text_cols:
            plt.xticks(x, df[text_cols[0]], rotation=45, ha='right')
        plt.ylabel('Value', fontsize=12)
        plt.xlabel(text_cols[0].replace('_', ' ').title() if text_cols else 'Category', fontsize=12)
        plt.legend()
        plt.grid(axis='y', alpha=0.3, linestyle='--')
    
    elif len(numeric_cols) == 1 and num_rows <= 8:
        # Case: Small dataset with categories - pie chart
        plt.figure(figsize=(10, 8))
        
        if text_cols:
            labels = df[text_cols[0]]
        else:
            labels = df.index
        
        colors = sns.color_palette("Set2", num_rows)
        plt.pie(df[numeric_cols[0]], labels=labels, autopct='%1.1f%%', 
                startangle=90, colors=colors, textprops={'fontsize': 11, 'fontweight': 'bold'})
    
    else:
        # Default: simple bar chart
        plt.figure(figsize=(12, 6))
        plt.bar(range(num_rows), df[numeric_cols[0]] if numeric_cols else df.iloc[:, -1], 
                color='steelblue', edgecolor='black')
        plt.xticks(range(num_rows), rotation=45, ha='right')
        plt.grid(axis='y', alpha=0.3, linestyle='--')
    
    # Set title
    if title:
        plt.title(title, fontsize=16, fontweight='bold', pad=20)
    else:
        plt.title(f'Query-{query_num} Results', fontsize=16, fontweight='bold', pad=20)
    
    plt.tight_layout()
    plt.show()
    print(f"âœ“ Visualization complete!\n")

# ========================================
# EXECUTE QUERY-1
# ========================================
print("\n" + "="*60)
print("QUERY-1: Content Watched by Subscriber")
print("="*60)

db = get_connection()
with open('Query-1.sql', 'r') as file:
    query1 = file.read()

result1 = pd.read_sql(query1, db)
db.close()

print(f"âœ“ Returned {len(result1)} rows")
print("Columns:", result1.columns.tolist())
print("\nFirst 10 rows:")
print(result1.head(10))

create_smart_visualization(result1, 1, 'Content Watched by Subscriber')

# ========================================
# EXECUTE QUERY-2
# ========================================
print("\n" + "="*60)
print("QUERY-2")
print("="*60)

db = get_connection()
with open('Query-2.sql', 'r') as file:
    query2 = file.read()

result2 = pd.read_sql(query2, db)
db.close()

print(f"âœ“ Returned {len(result2)} rows")
print("Columns:", result2.columns.tolist())
print("\nFirst 10 rows:")
print(result2.head(10))

create_smart_visualization(result2, 2)

# ========================================
# EXECUTE QUERY-3
# ========================================
print("\n" + "="*60)
print("QUERY-3")
print("="*60)

db = get_connection()
with open('Query-3.sql', 'r') as file:
    query3 = file.read()

result3 = pd.read_sql(query3, db)
db.close()

print(f"âœ“ Returned {len(result3)} rows")
print("Columns:", result3.columns.tolist())
print("\nFirst 10 rows:")
print(result3.head(10))

create_smart_visualization(result3, 3)

# ========================================
# EXECUTE QUERY-4
# ========================================
print("\n" + "="*60)
print("QUERY-4")
print("="*60)

db = get_connection()
with open('Query-4.sql', 'r') as file:
    query4 = file.read()

result4 = pd.read_sql(query4, db)
db.close()

print(f"âœ“ Returned {len(result4)} rows")
print("Columns:", result4.columns.tolist())
print("\nFirst 10 rows:")
print(result4.head(10))

create_smart_visualization(result4, 4)

# ========================================
# EXECUTE QUERY-5
# ========================================
print("\n" + "="*60)
print("QUERY-5")
print("="*60)

db = get_connection()
with open('Query-5.sql', 'r') as file:
    query5 = file.read()

result5 = pd.read_sql(query5, db)
db.close()

print(f"âœ“ Returned {len(result5)} rows")
print("Columns:", result5.columns.tolist())
print("\nFirst 10 rows:")
print(result5.head(10))

create_smart_visualization(result5, 5)

# ========================================
# EXECUTE QUERY-6
# ========================================
print("\n" + "="*60)
print("QUERY-6")
print("="*60)

db = get_connection()
with open('Query-6.sql', 'r') as file:
    query6 = file.read()

result6 = pd.read_sql(query6, db)
db.close()

print(f"âœ“ Returned {len(result6)} rows")
print("Columns:", result6.columns.tolist())
print("\nFirst 10 rows:")
print(result6.head(10))

create_smart_visualization(result6, 6)

# ========================================
# EXECUTE QUERY-7
# ========================================
print("\n" + "="*60)
print("QUERY-7")
print("="*60)

db = get_connection()
with open('Query-7.sql', 'r') as file:
    query7 = file.read()

result7 = pd.read_sql(query7, db)
db.close()

print(f"âœ“ Returned {len(result7)} rows")
print("Columns:", result7.columns.tolist())
print("\nFirst 10 rows:")
print(result7.head(10))

create_smart_visualization(result7, 7)

# ========================================
# EXECUTE QUERY-8
# ========================================
print("\n" + "="*60)
print("QUERY-8")
print("="*60)

db = get_connection()
with open('Query-8.sql', 'r') as file:
    query8 = file.read()

result8 = pd.read_sql(query8, db)
db.close()

print(f"âœ“ Returned {len(result8)} rows")
print("Columns:", result8.columns.tolist())
print("\nFirst 10 rows:")
print(result8.head(10))

create_smart_visualization(result8, 8)

# ========================================
# EXECUTE QUERY-9
# ========================================
print("\n" + "="*60)
print("QUERY-9")
print("="*60)

db = get_connection()
with open('Query-9.sql', 'r') as file:
    query9 = file.read()

result9 = pd.read_sql(query9, db)
db.close()

print(f"âœ“ Returned {len(result9)} rows")
print("Columns:", result9.columns.tolist())
print("\nFirst 10 rows:")
print(result9.head(10))

create_smart_visualization(result9, 9)

print("\n" + "="*60)
print("âœ“ ALL 9 QUERIES COMPLETED WITH VISUALIZATIONS!")
print("="*60)
```
